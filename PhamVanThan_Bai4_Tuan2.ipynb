```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#Cau 1.
coffee = cv2.imread('/content/images/coffee.jpeg')
coffee = cv2.cvtColor(coffee, cv2.COLOR_BGR2RGB)
coffee = cv2.resize(coffee, (50, 50))

paper = cv2.imread('/content/images/paper.png')
paper = cv2.cvtColor(paper, cv2.COLOR_BGR2RGB)
paper = cv2.resize(paper, (50, 50))

astronous = cv2.imread('/content/images/astronous.jpeg')
astronous = cv2.cvtColor(astronous, cv2.COLOR_BGR2RGB)
astronous = cv2.resize(astronous, (50, 50))

horse = cv2.imread('/content/images/horse.jpeg')
horse = cv2.cvtColor(horse, cv2.COLOR_BGR2RGB)
horse = cv2.resize(horse, (50, 50))
```

```python
inter_linear1 = cv2.resize(coffee, None, fx = 4, fy = 4)
inter_cubic1 = cv2.resize(coffee, None, fx = 4, fy = 4, interpolation = cv2.INTER_CUBIC)
inter_nearest1 = cv2.resize(coffee, None, fx = 4, fy = 4, interpolation = cv2.INTER_NEAREST)
inter_area1 = cv2.resize(coffee, (400,400), interpolation = cv2.INTER_AREA)
inter_lanczos41 = cv2.resize(coffee, (400,400), interpolation = cv2.INTER_LANCZOS4)

inter_linear2 = cv2.resize(paper, None, fx = 4, fy = 4)
inter_cubic2 = cv2.resize(paper, None, fx = 4, fy = 4, interpolation = cv2.INTER_CUBIC)
inter_nearest2 = cv2.resize(paper, None, fx = 4, fy = 4, interpolation = cv2.INTER_NEAREST)
inter_area2 = cv2.resize(paper, (400,400), interpolation = cv2.INTER_AREA)
inter_lanczos42 = cv2.resize(paper, (400,400), interpolation = cv2.INTER_LANCZOS4)

inter_linear3 = cv2.resize(astronous, None, fx = 4, fy = 4)
inter_cubic3 = cv2.resize(astronous, None, fx = 4, fy = 4, interpolation = cv2.INTER_CUBIC)
inter_nearest3 = cv2.resize(astronous, None, fx = 4, fy = 4, interpolation = cv2.INTER_NEAREST)
inter_area3 = cv2.resize(astronous, (400,400), interpolation = cv2.INTER_AREA)
inter_lanczos43 = cv2.resize(astronous, (400,400), interpolation = cv2.INTER_LANCZOS4)

inter_linear4 = cv2.resize(horse, None, fx = 4, fy = 4)
inter_cubic4 = cv2.resize(horse, None, fx = 4, fy = 4, interpolation = cv2.INTER_CUBIC)
inter_nearest4 = cv2.resize(horse, None, fx = 4, fy = 4, interpolation = cv2.INTER_NEAREST)
inter_area4 = cv2.resize(horse, (400,400), interpolation = cv2.INTER_AREA)
inter_lanczos44 = cv2.resize(horse, (400,400), interpolation = cv2.INTER_LANCZOS4)

Titles = ["Original 50x50", "Linear 400x400", "Cubic 400x400", "Nearest 400x400", "Area 400x400", "Lanczos4 400x400"]
Images1 = [coffee, inter_linear1, inter_cubic1, inter_nearest1, inter_area1, inter_lanczos41]
Images2 = [paper, inter_linear2, inter_cubic2, inter_nearest2, inter_area2, inter_lanczos42]
Images3 = [astronous, inter_linear3, inter_cubic3, inter_nearest3, inter_area3, inter_lanczos43]
Images4 = [horse, inter_linear4, inter_cubic4, inter_nearest4, inter_area4, inter_lanczos44]

fig, axes = plt.subplots(4, 6, figsize = (12,8))
for i in range(6):
    axes[0,i].imshow(Images1[i])
    axes[0,i].set_title(Titles[i])
    axes[0,i].axis('off')
for i in range(6):
    axes[1,i].imshow(Images2[i])
    axes[1,i].axis('off')
    axes[2,i].imshow(Images3[i])
    axes[2,i].axis('off')
    axes[3,i].imshow(Images4[i])
    axes[3,i].axis('off')
plt.subplots_adjust(wspace = 0, hspace = 0)
plt.show()
```
